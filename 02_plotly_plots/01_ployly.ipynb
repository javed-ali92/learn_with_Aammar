# plots on titanic Datasets

```python
# import dataset 
df = sns.load_dataset('titanic')
# check the head of the dataset
df.head()
```

```python
# scatter plot
fig = px.scatter(df, x='age', y='fare', color='sex', title='Plot of Titanic Dataset')
fig.show()
```

```python
# scatter plot
df.dropna(inplace=True)
fig = px.scatter(df, x='age', y='fare', color='sex', size= 'fare', title='Plot of Titanic Dataset')
fig.show()
```

```python
# scatter plot
#df.dropna(inplace=True)
fig = px.scatter(df, x='age', y='fare', color='sex', size= 'age', title='Plot of Titanic Dataset')
fig.show()
```

```python
# scatter plot
#df.dropna(inplace=True)
fig = px.scatter(df, x='age', y='fare', color='sex', size= 'age', title='Plot of Titanic Dataset')
fig.show()

# save the plot as a pmg image with 300dpi resolution
fig.write_image('./output/02_scatterplot.png', format='png', width=800, height=600, scale=3)
```

```python
import plotly.graph_objects as go

# create a simple line chart
fig = go.Figure(data=go.Scatter(x=[1, 2, 3, 4], y=[10, 11, 12, 13]))
fig.show()
```

```python
# customize the plot
fig.update_layout(
    title= 'Line Plot of dummy data',
    xaxis = dict(title='Number'),
    yaxis = dict(title='Value'),
showlegend = True,)
```

```python
# heatmape in plotly

data = np.random.rand(10,10)

heatmap = go.Heatmap(z=data)
fig = go.Figure(data=[heatmap])
fig.show()
```

```python
# heatmap on corr of titanic dataset of age and fare using plotly

fig = px.imshow(df[['age','fare']].corr())
fig.show()
```
