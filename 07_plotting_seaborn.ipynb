```python
# style of plot
sns.set_theme(style='darkgrid')
# scatter plot size of fig size
plt.figure(figsize=(8,4))

# change the scstter point colors
palette = {'First': 'blue', 'Second': 'red', 'Third': 'green'}
sns.boxplot(data = df, x= 'sex', y= 'age', hue = 'class', palette=palette)

# labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot Age vs Fare')

# change legend position and title
plt.legend(loc="upper right", title='class name')
```

```python
# style of plot
sns.set_theme(style='darkgrid')
# scatter plot size of fig size
plt.figure(figsize=(12,6))

# change the scstter point colors
palette = {'First': 'blue', 'Second': 'red', 'Third': 'green'}
sns.boxenplot(data = df, x= 'sex', y= 'age', hue = 'class', palette=palette)

# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxen Plot Age, based on Gender and Class name of passenger')

# change legend position and title
plt.legend(loc="upper right", title='class name')

# change the x-ticks name
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])

# rotate x ticks labels by 45 degree
plt.xticks(rotation=45)
```

```python
# bar plot using seaborn

# style of plot
sns.set_theme(style='darkgrid')
# scatter plot size of fig size
plt.figure(figsize=(12,6))

# change the scstter point colors
palette = {'First': 'blue', 'Second': 'red', 'Third': 'green'}
sns.barplot(data = df, x= 'sex', y= 'age', hue = 'class', palette=palette, capsize=0.1) # add capsize

# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Bar Plot Age, based on Gender and Class name of passenger')

# change legend position and title
plt.legend(loc="upper right", title='class name')

# change the x-ticks name
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])

# rotate x ticks labels by 45 degree
plt.xticks(rotation=45)
```

```python
# Violinplot using seaborn

# style of plot
sns.set_theme(style='darkgrid')
# scatter plot size of fig size
plt.figure(figsize=(12,6))

# change the scstter point colors
palette = {'First': 'blue', 'Second': 'red', 'Third': 'green'}
sns.violinplot(data = df, x= 'sex', y= 'age', hue = 'class', palette=palette)

# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Violin Plot Age, based on Gender and Class name of passenger')

# change legend position and title
plt.legend(loc="upper right", title='class name')

# change the x-ticks name
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])

# rotate x ticks labels by 45 degree
plt.xticks(rotation=45)
```
